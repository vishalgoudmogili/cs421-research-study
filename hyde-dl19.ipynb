```python
# Import necessary modules for the reproducibility
from pyserini.search import FaissSearcher, LuceneSearcher  # Import Faiss and Lucene searchers
from pyserini.search.faiss import AutoQueryEncoder  # Import AutoQueryEncoder for FaissSearcher
from pyserini.search import get_topics, get_qrels  # Import functions to get topics and qrels
from tqdm import tqdm  # Import tqdm for progress tracking
```

### Initialize Contriever Index and Query Encoder
We use [Pyserini](https://github.com/castorini/pyserini) as the search interface for the experiment. Please following the guidance in Pyserini to create Contriever index using the checkpoint from original Contriever work.

```python
# Create an AutoQueryEncoder using the 'facebook/contriever' model for encoding queries with mean pooling
query_encoder = AutoQueryEncoder(encoder_dir='facebook/contriever', pooling='mean')

# Initialize a FaissSearcher with the Faiss index created using 'contriever_msmarco_index/'
searcher = FaissSearcher('contriever_msmarco_index/', query_encoder)

# Create a LuceneSearcher using a prebuilt index for the MSMARCO passage corpus ('msmarco-v1-passage')
corpus = LuceneSearcher.from_prebuilt_index('msmarco-v1-passage')
```

### Load query and judegments for dl19-passage dataset

```python
# Retrieve topics for the DL19-passage dataset
topics = get_topics('dl19-passage')

# Retrieve query relevance judgments (qrels) for the DL19-passage dataset
qrels = get_qrels('dl19-passage')
```

## Run Contriever

```python
# Write retrieval results for DL19-passage using the 'contriever' model
with open('dl19-contriever-top1000-trec', 'w') as f:
    # Iterate through each query ID in topics
    for qid in tqdm(topics):
        # Check if the query ID is present in relevance judgments (qrels)
        if qid in qrels:
            # Extract the query from topics based on the query ID
            query = topics[qid]['title']
            
            # Perform a search using the 'contriever' model, retrieving the top 1000 hits
            hits = searcher.search(query, k=1000)
            rank = 0
            
            # Write the retrieval results to the output file
            for hit in hits:
                rank += 1
                f.write(f'{qid} Q0 {hit.docid} {rank} {hit.score} rank\n')
```

```python
import openai
import os
from openai.error import RateLimitError

# Define a function to call the Codex read API
def call_codex_read_api(prompt: str):
    # Define a function to parse the API result
    def parse_api_result(result):
        to_return = []
        for idx, g in enumerate(result['choices']):
            text = g['text']
            logprob = sum(g['logprobs']['token_logprobs'])
            to_return.append((text, logprob))
        res = [r[0] for r in sorted(to_return, key=lambda tup: tup[1], reverse=True)]
        return res
    
    try:
        # Make a request to the Codex read API
        result = openai.Completion.create(
            engine='text-davinci-003',
            prompt=prompt,
            api_key=os.getenv()['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'],
            max_tokens=512,
            temperature=0.7,
            top_p=1,
            n=8,
            stop=['\n\n\n'],
            logprobs=1
        )
        # Parse and return the API result
        return parse_api_result(result)
    
    except RateLimitError as e:
        # Handle rate limit error
        return f"Rate limit error: {e}"
```

## Run HyDE

```python
from time import sleep
import numpy as np
import json
from tqdm import tqdm

# Open files for writing
with open('hyde-dl19-contriever-gpt3-top1000-8rep-trec', 'w') as f, open('hyde-dl19-gpt3-gen.jsonl', 'w') as fgen:
    # Loop through topics
    for qid in tqdm(topics):
        if qid in qrels:
            # Get the query from topics
            query = topics[qid]['title']
            print(query)
            
            # Define the prompt for Codex read API
            prompt = f"""Please write a passage to answer the question
                        Question: {query}
                        Passage:"""
            
            get_result = False
            while not get_result:
                try:
                    # Call Codex read API to get contexts
                    contexts = [c.strip() for c in call_codex_read_api(prompt)] + [query]
                    
                    # Calculate embeddings for contexts
                    all_emb_c = [np.array(query_encoder.encode(c)) for c in contexts]
                    all_emb_c = np.array(all_emb_c)
                    avg_emb_c = np.mean(all_emb_c, axis=0)
                    avg_emb_c = avg_emb_c.reshape((1, len(avg_emb_c)))
                    
                    # Write to the generated file
                    fgen.write(json.dumps({'query_id': qid, 'query': query, 'contexts': contexts}) + '\n')
                    
                    get_result = True
                except:
                    sleep(1)
            
            print(contexts)
            
            # Perform a search using the embeddings
            hits = searcher.search(avg_emb_c, k=1000)
            
            # Write the search results to the TREC file
            rank = 0
            for hit in hits:
                rank += 1
                f.write(f'{qid} Q0 {hit.docid} {rank} {hit.score} rank\n')
```
